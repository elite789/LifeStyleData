```python
group1_cols = ['Gender', 'Wokout_Type', 'meal_name', 'meal_type']
group2_cols = ['diet_type', 'cooking_method', 'Workout_Type', 'Name of Exercise', 'Benefit']
group3_cols = ['Target Muscle Group', 'Equipment Needed', 'Difficulty Level', 'Type of Muscle', 'Workout', 'Body Part', 'Burns_Calories_Bin']
```

```python
def analyze_group(df, columns, group_name):

    print(f"\n{'='*80}")
    print(f"ANALYSIS - {group_name}")
    print(f"{'='*80}\n")

    sns.set(style='darkgrid', font_scale=1.2)

    for col in columns:
        temp_series = df[col].fillna('Missing')
        value_counts = temp_series.value_counts()

        #Viz
        fig, axes = plt.subplots(1,2, figsize=(16,8))
        fig.suptitle(f"Distribution of '{col}'", fontsize=16, fontweight='bold')

        sns.countplot(
            data=df, x=col, order=value_counts.index, palette='PuOr',
            ax=axes[0], edgecolor='black', legend=False, hue=col)

        axes[0].set_title(f"count of '{col}'", fontsize=12, fontweight='semibold')
        axes[0].tick_params(axis='x', rotation=45)
        axes[0].grid(axis='y', linestyle='--', alpha=0.6)

        for p in axes[0].patches:
            axes[0].annotate(
                f"{int(p.get_height())}",
                (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom', fontsize=10)
        
        value_counts.plot.pie(
            autopct='%1.1f%%',
            ax=axes[1],
            colors=sns.color_palette('coolwarm', len(value_counts)),
            startangle=90, shadow=True, wedgeprops={'edgecolor': 'white'})

        axes[1].set_ylabel('')
        axes[1].set_title(f'Percentage share of {col}', fontsize=14)
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

group1_cols = ['Gender', 'Workout_Type', 'meal_type']
analyze_group(df, group1_cols, group_name="Group 1 — Demographics & Meals")

```

```python
exclude = ['Name of Exercise', 'Benefit']
analyze_group(df, [c for c in group2_cols if c not in exclude], group_name="Group 2 — Dietary Data")
```

```python
analyze_group(df, group3_cols, group_name="Group 3 — Exercise Detail")
```
